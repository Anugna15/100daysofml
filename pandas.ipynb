```python
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies[movies.duration>=200].head(10)
```

```python
movies[(movies.duration>=200) & (movies.genre=='Action')].head(10)
```

```python
movies[(movies.duration>=200) | (movies.genre=='Action')].head(10)
```

```python
movies[(movies.duration>=200) & (movies.genre=='Action')|(movies.genre=='Drama')].head(10)
```

```python
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')].head(10)
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```
