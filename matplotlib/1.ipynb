```python
import matplotlib.pyplot as plt
x=[1,2,5,6,7,8]
y=[3,3,5,7,3,1]
plt.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.savefig('line_plot.png')
plt.show()
```

```python
plt.plot(x, y,color='violet')
plt.savefig('line_plot.png')
plt.show()
```

```python
plt.plot(x, y,linestyle='--')
plt.savefig('line_plot.png')
plt.show()
```

```python
plt.plot(x, y,marker='o')
plt.savefig('line_plot.png')
plt.show()
```

```python
plt.plot(x, y,color='pink',linestyle='--',marker='s',linewidth=2)
plt.show()
```

```python
plt.plot(x, y,'g--o')
plt.show()
```

```python
styles = ['r--o', 'b-.s', 'g:*']
for i in range(3):
 plt.plot(x, [v * (i+1) for v in y], styles[i])
```

```python
x = [1, 2, 3, 4,4,3,2,1]
y = [2, 4, 6, 8,1,1,1,1]
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y)
plt.title("customized plot",fontsize=12,color='red')
plt.xlabel('input')
plt.ylabel("output")
plt.show()
```

```python
plt.plot(x, y, label='y = 2x')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='y = 2x')
plt.legend(loc='upper right')
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.figure(figsize=(8, 4)) # Width=8 inches, Height=4 inches
plt.plot(x, y)
plt.title('Custom Figure Size')
plt.show()
```

```python
plt.figure(figsize=(6, 3), dpi=150)
plt.plot(x, y)
plt.title('Higher DPI')
plt.show()
```

```python
plt.plot(x, y)
plt.title('High Quality Transparent')
plt.savefig('clean_plot.png', dpi=300, transparent=True)
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 7]
plt.subplot(1, 2, 1) # (rows, cols, index)
plt.plot(x, y1)
plt.title('Square')
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Prime')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 4)) # 1 row, 2 columns
axs[0].plot(x, y1)
axs[0].set_title('Square')
axs[1].plot(x, y2)
axs[1].set_title('Prime')
plt.tight_layout()
plt.show()
```

```python
plt.plot(x, y)
plt.title('Custom Ticks')
plt.xticks([1, 2, 3, 4], ['1s', '2s', '3s', '4s']) # Custom x-tick labels
plt.yticks([2, 4, 6, 8], ['2 meters', '4 meters', '6 meters', '8 meters']) # Custom y-tick labels
plt.show()
```

```python
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
values = [10, 15, 7, 10, 20]
plt.plot(labels, values)
plt.xticks(rotation=45) # Rotates x-axis labels by 45 degrees
plt.show()
```

```python
plt.plot(x, y)
plt.title('Custom Axis Limits')
plt.xlim(0, 5) # Set x-axis from 0 to 5
plt.ylim(0, 10) # Set y-axis from 0 to 10
plt.show()
```

```python
plt.plot(x, y)
plt.title('Logarithmic Scale')
plt.xscale('log') # Set x-axis to logarithmic scale
plt.yscale('log') # Set y-axis to logarithmic scale
plt.show()
```
